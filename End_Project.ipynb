```python

team_counts = df['Team'].value_counts()
percentages = (team_counts / team_counts.sum()) * 100

plt.figure(figsize=(10, 8))
bars =sns.barplot(x=team_counts.values, y=team_counts.index, color='#01F9C6')
plt.title('Employee Distribution by Team')
plt.xlabel('Count')
plt.ylabel('Teams')
plt.xticks(rotation=90)

plt.show()

#Height employee distribution by a team is New Orleans Pelicans and Memphis Grizzlies .
```

```python

custum_colors = ['#33afff','#ffa233','#ffe933','#2cb26f','#752cb2']

plt.figure(figsize=(10, 6))
position_counts = df['Position'].value_counts()
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', colors =custum_colors )
plt.title('Position Distribution')
plt.show()
print (f'Total number of  Position Distribution is:{position_counts}')
```

```python
bins=[20,25,30,35,40]
labels = ['20-25','26-30','31-40']
age_group_count = df['Age'].value_counts()
plt.figure(figsize=(8,5))
sns.histplot(df['Age'],bins = 5,kde = True)
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

# In age group distribution the age between 20-25 have more employees comparing with other group.
```

```python
salary_exp = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
max_salary = salary_exp.loc[salary_exp['Salary'].idxmax()]
print(max_salary)

```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data = salary_exp,x='Salary',y='Team',hue='Position')
plt.title('Salary Expenditure')
plt.show()
        
```

```python
correlation= df[['Age','Salary']].corr()
print(correlation)
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(data = correlation,annot=True,cmap ='mako',linewidth = 1.0)
plt.title('Age vs Salary')
plt.show()
```
